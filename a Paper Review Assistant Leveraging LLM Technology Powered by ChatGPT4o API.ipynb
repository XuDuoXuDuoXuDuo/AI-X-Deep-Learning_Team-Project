```python
!pip install openai==0.28
!pip install pdfplumber
import os
import openai
import pdfplumber
from google.colab import files

## Set up your OpenAI API key
# 设置您的 OpenAI API 密钥
openai.api_key = "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"

# Upload PDF file to Colab
# 上传 PDF 文件到 Colab
uploaded = files.upload()  # 手动上传 PDF 文件 Manually upload PDF files
file_paths = []

for file_name in uploaded.keys():
    if file_name.endswith(".pdf"):
        file_path = os.path.join("/content", file_name)
        file_paths.append(file_path)
        print(f"Uploaded {file_name} to path {file_path}")

# Extract the text content of the PDF file
# 提取 PDF 文件的文本内容
def extract_text_from_pdfs(file_paths):
    all_text = ""
    char_count = 0
    max_chars = 300000  # 设置最大字符数量为 300000 Set the maximum number of characters to 300000
    for file_path in file_paths:
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                if char_count >= max_chars:
                    break
                text = page.extract_text()
                if text:
                    remaining_chars = max_chars - char_count
                    all_text += text[:remaining_chars] + ""
                    char_count += len(text[:remaining_chars])
    return all_text

# Extract the text content of all PDF files
# 提取所有 PDF 文件的文本内容
pdf_text = extract_text_from_pdfs(file_paths)

# Save the extracted text to a TXT file for download
# 将提取的文本保存到一个 TXT 文件中供下载
output_text_file = "/content/extracted_pdf_text.txt"
with open(output_text_file, "w") as text_file:
    text_file.write(pdf_text)

print(f"Extracted text saved to {output_text_file}")

# Provide download link
# 提供下载链接
from google.colab import files
files.download(output_text_file)

# Function used to create a conversation and get a reply
# 函数用于创建对话并获取回复
def create_thread_and_run(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-4o",  # 使用最新的 GPT-4o 模型 Use the latest GPT-4o model
        messages=[
            {"role": "user", "content": prompt}
        ],
        max_tokens=10000,
        temperature=0.7
    )
    return response

# Print the Assistant's response
# 打印 Assistant 的响应
def pretty_print(response):
    print("# Assistant Response")
    print(response['choices'][0]['message']['content'].strip())

# Set the prompt to include the extracted PDF text content
# 设定 prompt，将提取的 PDF 文本内容包含进去
prompt = f"""
The key aspects of foodborne virus culture technology using zebrafish include:
1. Background and advantages of using zebrafish as a model, including comparison with other animal models
2. Classification and impact of foodborne viruses on global food safety
3. Technical details of viral culture in zebrafish, including infection methods
4. Viral replication dynamics and zebrafish immune response
5. Influence of environmental factors, like oxygen levels, on viral infection in zebrafish
6. Role of gut microbiota in viral replication within zebrafish
7. Application of zebrafish for antiviral drug screening and toxicity testing
8. Limitations of zebrafish as a model and future research directions.

Here is the extracted content from the provided PDF files:

{pdf_text}

Based on this content, please summarize the key aspects and provide insights for each of these topics.
"""
# Run the Assistant and print the response
# 运行 Assistant 并打印响应
response = create_thread_and_run(prompt)
pretty_print(response)
```
